# Ataque por cercanía

Importamos librerías 

```python
#Para crear y análizar redes
import networkx as nx
#Para dibujar redes y graficar funciones
import matplotlib.pyplot as plt
#Para leer desde un archivo
import pandas as pd 
import numpy as np
#Para generar errores aleatorios
import random
```

# # Función de Mapa de calor por cercanía.

```python
def mapa_calor_cercania(G, nombre_mcalor_grado):
    degree_map = {node:nx.closeness_centrality(G, node) for node in G.nodes()}
    #valores máximo y mínimos de grado
    max_degree = max(degree_map.values())
    min_degree = min(degree_map.values())
    #el grado nodal nos da el color de nodo, el cual se mapea con el mapa de color
    colores_nodos = list(degree_map.values())
    #ver mapas de color en https://matplotlib.org/stable/tutorials/colors/colormaps.html
    mapa_color = plt.get_cmap('spring')
    #dibujamos la red
    nx.draw(G, cmap=mapa_color, node_color=colores_nodos, with_labels=True)
    #dibujamos la barra de color
    sm = plt.cm.ScalarMappable(cmap=mapa_color,norm=plt.Normalize(vmin = min_degree, vmax=max_degree))
    cbar = plt.colorbar(sm)
    cbar.set_label('Cercanía')
    plt.savefig(nombre_mcalor_grado + ".png")
    plt.show()
    # la siguiente instruccion limpia la figura, evitando que al graficar varias figuras, estas no se encimen 
    plt.clf()
```

## Función para obtener el número de componentes conectados.

```python
def num_componentes_conectados(G):
    num_cc=nx.number_connected_components(G)
    print("Numero de componentes conectados:"+str(num_cc))
    return num_cc
```

## Función para obtener la proporción del componente Gigante.

```python
def prop_componente_gigante(G):
    num_cc=num_componentes_conectados(G)
    Gcc =sorted(nx.connected_components(G), key=len, reverse=True)
    GC= G.subgraph(Gcc[0])
    n=len(nx.nodes(G))
    n_GC=len(nx.nodes(GC))
    pcg=n_GC/n
    print("La proporcion nodal entre Componente gigante y Red:"+str(pcg))
    return pcg
```

# #Función para obtener el nodo con mayor cercanía.

```python
def mayor_cercania(G):
    cercania=nx.closeness_centrality(G)
    cercania_mayor=max(cercania.values())
    print("Mayor cercania:"+str(cercania_mayor))
    for key in cercania:
        if cercania_mayor==cercania[key]:
            print("Nodo #"+str(key)+" Mayor cercania")
            return key
```

## Función graficadora

```python
def grafica_funciones(titulo,f1_valores, f2_valores,f1_leyenda, f2_leyenda ,ejex_etiqueta, ejey_etiqueta):
    plt.title(titulo)#se agrega un titulo a la gráfica
    plt.plot(f1_valores, label=f1_leyenda)#se ingresa la lista de valores a graficar con su respectiva etiqueta
    plt.plot(f2_valores, label=f2_leyenda)#se ingresa la lista de valores a graficar con su respectiva etiqueta
    plt.xlabel(ejex_etiqueta)#se agrega una etiqueta al eje x
    plt.ylabel(ejey_etiqueta)#se agrega una etiqueta al eje y
    plt.legend()#agrega una leyenda a la gráfica que se sobrepondrá con los nombres de las etiquetas de los valores
```

# #Función para atacar nodo con mayor cercanía.

```python
def ataque_cercania(G):
    nodo_atacado=mayor_cercania(G)
    G.remove_node(nodo_atacado)
    print("Nodo #"+str(nodo_atacado)+" fue atacado")
    return G
```

## Función para simulara varios ataques por cercanía.

```python
def ataques_secuenciales(G,num_ataques):
    num_cc_lista=list()
    prop_gc_lista=list()
    for i in range(num_ataques):
        red_atacada=ataque_cercania(G)
        print("Ataque número: "+str(i+1))
        num_cc_lista.append(num_componentes_conectados(red_atacada))
        prop_gc_lista.append(prop_componente_gigante(red_atacada))
    return (num_cc_lista, prop_gc_lista)
```

##  Creación de red de escala libre

```python
free_scale=nx.barabasi_albert_graph(100,5)
```

## Creación de red aleatoria

```python
aleatoria_red=nx.erdos_renyi_graph(100,0.1)
```

# Mapa de calor por cercanía.


```python
mapa_calor_cercania(free_scale,"Cercania_escalalibre")
```

```python
mapa_calor_cercania(aleatoria_red,"Cercania_aleatoria")
```

## Simulación de ataques por cercanía en la red de escala libre.

```python
free_scale_test = free_scale.copy()
#num_cc_lista=[]
#prop_gc_lista=[]
(num_cc_lista_free_scale,prop_gc_lista_free_scale)=ataques_secuenciales(free_scale_test, 99)
```

## Simulación de ataques por cercanía en la red de escala libre.

```python
aleatoria_red_test = aleatoria_red.copy()
#num_cc_lista=[]
#prop_gc_lista=[]
(num_cc_lista_aleatoria,prop_gc_lista_aleatoria)=ataques_secuenciales(aleatoria_red_test, 99)
```

## Gráfica de los componentes conectados vs numero de errores

```python
titulo="Componentes Vs Numero de Errores"
f1_valores=num_cc_lista_free_scale
f2_valores=num_cc_lista_aleatoria
f1_leyenda="Escala Libre"
f2_leyenda="Red Aleatoria"
ejex_etiqueta="Errores"
ejey_etiqueta="Componentes"
grafica_funciones(titulo,f1_valores, f2_valores,f1_leyenda, f2_leyenda ,ejex_etiqueta, ejey_etiqueta)
```
